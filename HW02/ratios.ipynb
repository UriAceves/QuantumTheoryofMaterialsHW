```julia
E_a = 10.
E_b = 1.

Δ = (E_a - E_b)/2

points = 1000

x_range = linspace(0.09,7.,points)

β_array = zeros(x_range)
E_array = zeros(x_range)
ratios = zeros(x_range)
domain = zeros(x_range)
for i in 1:points
    β_array[i] = x_range[i]
    E_array[i] = (E_a + E_b)/2 + sqrt(Δ^2 + β_array[i]^2)
    γ = (E_a - E_array[i])/β_array[i]
    c_a = 1/sqrt(1+γ^2)
    c_b = -γ*c_a
    ratios[i] = (c_a/c_b)^2
    domain[i] = Δ/β_array[i]
end

plot(domain,ratios)
xlabel(L"\Delta/\beta")
ylabel(L"(c_a/c_b)^2")
title("Anti-bonding state with "*L"$E_A = 10.0, E_B=1.0$")
savefig("anti-10-mal.png")
show()
```

## Spectrum

```julia
N = [5,10,50,100,200]

α = 1.
β = 0.1*α

diag = ones(N[1])
off_diag = β.*ones(N[1]-1)

H = Array(Tridiagonal(off_diag,diag,off_diag))
spectrum = eigvals(H)
plot(ones(N[1]),spectrum, "k_")

diag = ones(N[2])
off_diag = β.*ones(N[2]-1)

H = Array(Tridiagonal(off_diag,diag,off_diag))
spectrum = eigvals(H)
plot(2.*ones(N[2]),spectrum, "k_")

diag = ones(N[3])
off_diag = β.*ones(N[3]-1)

H = Array(Tridiagonal(off_diag,diag,off_diag))
spectrum = eigvals(H)
plot(3.*ones(N[3]),spectrum, "k_")

diag = ones(N[4])
off_diag = β.*ones(N[4]-1)

H = Array(Tridiagonal(off_diag,diag,off_diag))
spectrum = eigvals(H)
plot(4.*ones(N[4]),spectrum, "k_")

diag = ones(N[5])
off_diag = β.*ones(N[5]-1)

H = Array(Tridiagonal(off_diag,diag,off_diag))
spectrum = eigvals(H)
plot(5.*ones(N[5]),spectrum, "k_")
xticks([1,2,3,4,5],N)
xlabel("N")
ylabel(L"E in $\alpha$ units")
title("Eigenenergies of a finite atom chain")
savefig("finite_chain.png")
show()
```

```julia
N = [5,10,50,100,200]

α = 1.
β = 0.1*α

diag = ones(N[1])
off_diag = β.*ones(N[1]-1)

H = Array(Tridiagonal(off_diag,diag,off_diag))
H[1,size(H)[1]] = β
H[size(H)[1],1] = β
spectrum = eigvals(H)
plot(ones(N[1]),spectrum, "k_")

diag = ones(N[2])
off_diag = β.*ones(N[2]-1)

H = Array(Tridiagonal(off_diag,diag,off_diag))
H[1,size(H)[1]] = β
H[size(H)[1],1] = β
spectrum = eigvals(H)
plot(2.*ones(N[2]),spectrum, "k_")

diag = ones(N[3])
off_diag = β.*ones(N[3]-1)

H = Array(Tridiagonal(off_diag,diag,off_diag))
H[1,size(H)[1]] = β
H[size(H)[1],1] = β
spectrum = eigvals(H)
plot(3.*ones(N[3]),spectrum, "k_")

diag = ones(N[4])
off_diag = β.*ones(N[4]-1)

H = Array(Tridiagonal(off_diag,diag,off_diag))
H[1,size(H)[1]] = β
H[size(H)[1],1] = β
spectrum = eigvals(H)
plot(4.*ones(N[4]),spectrum, "k_")

diag = ones(N[5])
off_diag = β.*ones(N[5]-1)

H = Array(Tridiagonal(off_diag,diag,off_diag))
H[1,size(H)[1]] = β
H[size(H)[1],1] = β
spectrum = eigvals(H)
plot(5.*ones(N[5]),spectrum, "k_")
xticks([1,2,3,4,5],N)
xlabel("N")
ylabel(L"E in $\alpha$ units")
title("Eigenenergies of a periodic atom chain")
savefig("periodic_chain.png")
show()
```

```julia
finen(m,N) = α + 2*β*cos(m*π/(N+1))
finen2(m,N) = α + 2*β*cos(m*2*π/N)
```

```julia
N = [5,10,50,100,200]

α = 1.
β = 0.1*α

diag = ones(N[1])
off_diag = β.*ones(N[1]-1)

H = Array(Tridiagonal(off_diag,diag,off_diag))
spectrum = eigvals(H)

ana_ener = zeros(N[1])

for i in 1:N[1]
    ana_ener[i] = finen(i,N[1])
end

plot(ones(N[1]),spectrum, "k_")
plot(1.1*ones(N[1]),ana_ener, "k_")
tick_params(
    axis="x",          # changes apply to the x-axis
    which="both",      # both major and minor ticks are affected
    bottom="off",      # ticks along the bottom edge are off
    top="off",         # ticks along the top edge are off
    labelbottom="off")
savefig("fin_comp.png")
show()
```
