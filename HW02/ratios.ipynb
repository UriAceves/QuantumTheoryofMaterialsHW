```julia
using PyPlot
```

```julia
E_a = 100.
E_b = 1.

Δ = E_a - E_b

points = 1000

x_range = linspace(0.1,5,points)

β_array = zeros(x_range)
E_array = zeros(x_range)
ratios = zeros(x_range)
domain = zeros(x_range)
for i in 1:points
    β_array[i] = x_range[i]*Δ
    E_array[i] = (E_a + E_b)/2 + sqrt(Δ^2 + β_array[i]^2)
    γ = (E_array[i]-E_a)/β_array[i]
    c_a = 1/sqrt(1+γ)
    c_b = γ*c_a
    ratios[i] = (c_a/c_b)^2
    domain[i] = Δ/β_array[i]
end

plot(domain,ratios)
xlabel(L"\Delta/\beta")
ylabel(L"(c_a/c_b)^2")
show()
```
