```julia
using PyPlot
```

```julia
D1(E::Float64) = (sqrt(2)/π)/sqrt(E)
D2(E::Float64) = 1/π
D3(E::Float64) = 0.5*((2^1.5)/(π^2))*sqrt(E)
```

```julia
points = 1000

E = linspace(0,10,points)

ar1 = Array([D1(i) for i in E])
ar2 = Array([D2(i) for i in E])
ar3 = Array([D3(i) for i in E]);
```

```julia
fig = figure("DOS for 1D, 2D, and 3D",figsize=(15,5))
subplot(131)
plot(E,ar1)
title("DOS for free electron gas in 1D")
xlabel(L"$\varepsilon$ (Ha)")
subplot(132)
plot(E,ar2)
title("DOS for free electron gas in 2D")
xlabel(L"$\varepsilon$ (Ha)")
subplot(133)
plot(E,ar3)
title("DOS for free electron gas in 3D")
xlabel(L"$\varepsilon$ (Ha)")
savefig("Dos.png")
show()
```
