```julia
delt(x) = 1000
```

```julia
function KPPot(x)
    if abs(x)%1 < 1e-4
        return delt(x)
    end
    return 0.
end
```

```julia
function step(x)
    if abs(x)%2 < 1
        return 1.
    else 
        return 0.
    end
end
```

```julia
domain = linspace(0,11,1101)
codomain = [KPPot(x) for x in domain];
```

```julia
domain
```

```julia
using PyPlot
```

```julia
plot(domain,codomain,"r.")
savefig("delta.png")
show()
```

```julia
plot(real(fft(codomain)))
show()
```

```julia
cod2 = [step(x) for x in domain]
cod2 = cod2[51:end]
domain = linspace(0,10,1051)
plot(domain,cod2)
```

```julia
ft = fft(cod2)
xlim([10,1000])
ylim([-100,100])
plot(real(ft))
```
