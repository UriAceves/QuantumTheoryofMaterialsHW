```julia
kf = 3.
```

```julia
function en(k)
    return (k^2/2) - (kf)/(2π)*(2 + ((kf^2 - k^2)/(k*kf))*log(abs((k+kf)/(k-kf))))
end
```

```julia
using PyPlot
```

```julia
domain = linspace(0.1,6,300)
```

```julia
codomain = [en(x) for x in domain];
```

```julia
plot(domain,codomain)
axvline(3.)
title("Energy dispersion")
xlabel("k")
ylabel("E(k) (Ha)")
savefig("dispEn.png")
show()
```

```julia
function n(d)
    return 0.5*m*(2- 9*(sin(kf*d) - kf*d*cos(kf*d))/(kf*d)^3)
end
```

```julia
codomain = [n(d) for d in domain];
```

```julia
plot(domain,codomain)
title("Exchange correlation hole")
xlabel("δ")
ylabel("n(δ)")
savefig("hole.png")
show()
```
